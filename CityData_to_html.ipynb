# Create HTML Table from the pandas dataframe

```python
#import dependencies
import pandas as pd
```

```python
#save file path
file = '../python-api-challenge/output_data/City_Weather_data.csv'

#read csv into a dataframe
city_weather_df = pd.read_csv(file)
```

```python
city_weather_df.reset_index(inplace = True)
city_weather_df.rename(columns={"index": "City_ID"}, inplace = True)
city_weather_df
```

```python
#convert df into html table
html_table = city_weather_df.to_html(justify='left', index=False)
```
